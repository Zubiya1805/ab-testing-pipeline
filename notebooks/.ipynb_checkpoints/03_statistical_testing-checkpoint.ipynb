```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('../data/raw/ab_test_data.csv')
```

```python
df.head()
```

```python
control = df[df['group'] == 'control']['converted']
treatment = df[df['group']=='treatment']['converted']
```

```python
print(len(control))
print(len(treatment))
```

```python
from statsmodels.stats.proportion import proportions_ztest

# Number of conversions in each group
control_conversions = control.sum()
treatment_conversions = treatment.sum()

# Number of users in each group
control_total = len(control)
treatment_total = len(treatment)

print(f"Control conversions: {control_conversions} out of {control_total}")
print(f"Treatment conversions: {treatment_conversions} out of {treatment_total}")
```

```python
counts = [treatment_conversions, control_conversions]
nobs = [treatment_total, control_total]

z_stat, p_value = proportions_ztest(counts, nobs)

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
control_rate = control.mean()
treatment_rate = treatment.mean()

uplift = (treatment_rate- control_rate)/control_rate*100

monthly_visitors = 1000000
extra_purchases = (treatment_rate - control_rate)*monthly_visitors
```

```python
print(f"Control conversion rate: {control_rate:.2%}")
print(f"Treatment conversion rate: {treatment_rate:.2%}")
print(f"Relative uplift: {uplift:.2f}%")
print(f"Extra purchases per month: {extra_purchases:,.0f}")
```
