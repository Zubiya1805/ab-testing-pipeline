```python
import pandas as pd
import numpy as np
np.random.seed(42)
n_users = 10000
control_size = n_users//2
treatment_size = n_users//2
```

```python
control_users = pd.DataFrame({
    'user_id':np.arange(1,control_size+1),
    'group':'control',
    'converted': np.random.binomial(1,0.10,control_size)
})

control_users.head()
```

```python
treatment_users = pd.DataFrame({
    'user_id':np.arange(control_size+1,n_users+1),
    'group':'treatment',
    'converted': np.random.binomial(1,0.12,treatment_size)
})

treatment_users.head()
```

```python
df = pd.concat([control_users,treatment_users],ignore_index = True)
df.head()
```

```python
df.to_csv('../data/raw/ab_test_data.csv', index=False)
```

```python
# Save processed version with additional columns
df['conversion_rate'] = df.groupby('group')['converted'].transform('mean')
df['is_control'] = (df['group'] == 'control').astype(int)

df.to_csv('../data/processed/ab_test_processed.csv', index=False)
print("Processed data saved!")
```
